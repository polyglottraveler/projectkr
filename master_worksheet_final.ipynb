## The Challenge

Microsoft sees all the big companies creating original video content, and they want to get in on the fun. They have decided to create a new movie studio, but the problem is they don’t know anything about creating movies. They have hired you to help them better understand the movie industry. Your team is charged with doing data analysis and creating a presentation that explores what type of films are currently doing the best at the box office. You must then translate those findings into actionable insights that the CEO can use when deciding what type of films they should be creating.


## The Goal & Questions to Answer

- Increase <b>awareness</b>
- Increase <b>revenue</b> (existing customers spend more & attract new customers)
- Create <b>synergies</b> with existing business line (servers, XBox)
- <b>Compete</b> with Amazon Prime Video, Apple TV

Given the background and goals, our questions of interest were:

- Q1: Which <b>studio</b> should Microsoft partner with to create video?
- Q2: What <b>genres</b> can generate most revenue, profit and profit margin?
- Q3: What are the <b>seasonalities</b> in movies industry?
- Q4: Are there any movie <b>categories</b> that generate the greatest worldwide revenue?
- Q5: Who are the <b>top 5 names</b> by worldwide revenue? Top actors/actresses?
- Q6: Do <b>TMDB ratings</b> indicate success?

Analyzing the data from Box Office Mojo, IMDB, Rotten Tomatoes and TheMovieDB.org to answer these questions will help them formulate Go-To-Market Strategies to launch original video content.

## Q2: What genres can generate most revenue, proft and profit margin?

```python
# import data

imdb_dataset = pd.read_csv('imdb_dataset.csv')
imdb_dataset

tn_dataset = pd.read_csv('tn_dataset.csv')
tn_dataset
```

```python
# merging these two tables to connect genres with gross revenue and product budget

imdb_tn = imdb_dataset.merge(tn_dataset, how = 'inner', left_on=['primary_title'], right_on=['movie'] )
imdb_tn
```

```python
# teasing out genres data

genres_new = imdb_tn['genres'].apply(lambda x: str(x).split(',')).apply(pd.Series)
genres_new
```

```python
# adding genres back to the merged table

imdb_tn_full = imdb_tn.merge(genres_new, left_index = True, right_index = True)
imdb_tn_full
```

```python
# removing the duplicated or unncessary columns

imdb_tn_full.drop(["genres", 'primary_title',"id", 'movie'], axis = 1, inplace=True)
imdb_tn_full
```

```python
# melting genres to multiple rows and dropping the rows with null genre

imdb_tn_full = imdb_tn_full.melt(id_vars = ['tconst', 'original_title', 'start_year', 'runtime_minutes','release_date', 'domestic_gross', 'worldwide_gross', 'production_budget'], value_name = 'genre')
imdb_tn_full.dropna(subset=['genre'], inplace = True)
imdb_tn_full = imdb_tn_full.sort_values(by = 'genre')
```

```python
# converting $ cells into numeric cells
imdb_tn_full[imdb_tn_full.columns[5:8]] = imdb_tn_full[imdb_tn_full.columns[5:8]].replace('[\$,]', '', regex=True).astype(float)
imdb_tn_full
```

```python
# calculating profit and and profit margin

imdb_tn_full['profit'] = round(imdb_tn_full['worldwide_gross'] - imdb_tn_full['production_budget'] )
imdb_tn_full['profit%'] = round(imdb_tn_full['profit'] / imdb_tn_full['production_budget'] * 100)
```

```python
# grouping dataset by genre

imdb_tn_full2 = imdb_tn_full.groupby(['genre'], as_index=False).mean()
imdb_tn_full2
```

```python
# plotting gross revenue by genre

sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize = (20,7))


pal = sns.color_palette("Blues_d", len(imdb_tn_full2.worldwide_gross))
rank = imdb_tn_full2.worldwide_gross.argsort().argsort()   
sns.barplot(x = imdb_tn_full2.genre, y= imdb_tn_full2.worldwide_gross/1e6, palette = np.array(pal[::-1])[rank])

ax.set_xlabel('Genre', fontsize=15)
ax.set_ylabel('Revenue in Millions', fontsize=15)
ax.set_title('Worldwide Revenue by Genre in Millions', fontsize=20)
plt.xticks(rotation='45')
```

### To generate revenue & awareness, prioritize Adventure, Animation, Fantasy, Musical and Sci-Fi genres


```python
#plotting gross revenue by genre

sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize = (20,7))

sns.barplot(x = imdb_tn_full.genre, y= imdb_tn_full.worldwide_gross/1e6, color = 'b')

ax.set_xlabel('Genre', fontsize=15)
ax.set_ylabel('Revenue in Millions', fontsize=15)
ax.set_title('Worldwide Revenue by Genre in Millions', fontsize=20)
plt.xticks(rotation='45')
```

### Be careful about Musical genre, it's either a hit or a miss

```python
# plotting profit by genre + average margin

sns.set(style="ticks")

fig, ax1 = plt.subplots(figsize = (20,7))
ax2 = ax1.twinx()

pal = sns.color_palette("Blues_d", len(imdb_tn_full2.profit))
rank = imdb_tn_full2.profit.argsort().argsort()   
sns.barplot(x = imdb_tn_full2.genre, y= imdb_tn_full2.profit/1e6, palette = np.array(pal[::-1])[rank], ax=ax1)
#sns.barplot(x = 'genre', y= "profit", data = imdb_tn_full2, label='profit', ax=ax1, color = 'b')
sns.lineplot(x = 'genre', y= 'profit%', data = imdb_tn_full2, label='profit%', ax=ax2, color = 'r')


# setting titles and axis labels
ax1.set_title('Profit (Millions) and Profitability by Genre', fontsize=20)
ax1.set_ylabel('Profit in Millions', fontsize=15)
# plt.xticks(rotation = 45)
ax2.set_ylabel('Profitability', fontsize=15)

for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
```

### To generate profit, Adventure, Animation, Musical and Sci-Fi still are the best bets, but profitability of Musical and Horror are the highest

## Q3: What are the seasonalities in movies industry?

```python
# import release date data

imdb_tn_full
```

```python
# converting release date to datetime and extract month

imdb_tn_full['release_date'] = pd.to_datetime(imdb_tn_full['release_date'])
imdb_tn_full['month'] = imdb_tn_full['release_date'].dt.month
imdb_tn_full['Month-str'] = imdb_tn_full['release_date'].dt.strftime('%b')
imdb_tn_full3 = imdb_tn_full.sort_values(by=['month'], ascending = True)
imdb_tn_full3
```

```python
# plotting lineplot revenue seasonality by month

sns.set(style="white")

fig, ax1 = plt.subplots(figsize = (15,6))
sns.lineplot(x=imdb_tn_full3.month, y=imdb_tn_full3.worldwide_gross/1e6, label='worldwide revenue (M)')
sns.lineplot(x=imdb_tn_full3.month, y= imdb_tn_full3.domestic_gross/1e6, label='domestic revenue (M)')

ax1.set_xlabel('Month', fontsize=15)
ax1.set_ylabel('Worldwide Revenue in Millions', fontsize=15)
ax1.set_title('Movie Seasonality by Month', fontsize=20)
```

### Summer (May, June, July) and holiday seasons (Nov) are the best to launch new titles

## Q4: Are there any movie categories that generate the greatest worldwide revenue?


```python
#Importing and final cleaning of data to get to a useful stage

# See kca_joined_rev.ipynb for how this table was made
df_joint_rev = pd.read_csv('joined_revenue')
# Taking top 50 revenue films at the box office
df_joint_rev_50_largest=df_joint_rev.nlargest(50,'worldwide_gross')
#See kca_rev_actor_movie_cat.ipynb for how this table was made 
df_top50_title_descr=pd.read_csv('movie_title_descr')
df_rev_top_50_cat = df_joint_rev_50_largest.merge(df_top50_title_descr, how='inner',left_on=['movie_title'], right_on=['movie_title_list'])
df_rev_top_50_cat_skinny = df_rev_top_50_cat.drop(['movie_title','production_budget','domestic_gross','release_year','movie_title_list'], axis=1)
df_top_50_cat_grouped = df_rev_top_50_cat_skinny.groupby(['movie_descr']).sum().sort_values(by=['worldwide_gross'], ascending = False)
df_top_50_cat_grouped = df_top_50_cat_grouped.reset_index()
```
